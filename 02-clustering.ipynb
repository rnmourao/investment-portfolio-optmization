```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_pickle('data/history.pkl')
```

```python
df.head()
```

```python
df[['PETR4.SA', 'BBAS3.SA']].plot();
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
dfs = pd.DataFrame(mms.fit_transform(df), columns=df.columns, index=df.index)
```

```python
dfs
```

```python
dfs[['PETR4.SA', 'BBAS3.SA']].plot();
```

```python
from sklearn.cluster import KMeans
```

```python
dft = dfs.transpose()
```

```python
dft[dft.isna().any(axis=1)]
```

```python
dft = dft.dropna()
```

```python
dft[dft.isna().any(axis=1)]
```

```python
ls = []
for n in range(2, 6):
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(X=dft)
    ls.append({'n_clusters': n, 'wcss': kmeans.inertia_})
elbow = pd.DataFrame(ls)
elbow.set_index('n_clusters', inplace=True)
elbow.head()
```

```python
elbow.plot();
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
clusters = pd.DataFrame(kmeans.fit_predict(dft), index=dft.index)
clusters.columns = ['cluster']
clusters.reset_index(inplace=True)
clusters.head()
```

```python
dft.transpose()[['AALR3.SA', 'ABEV3.SA']].plot();
```

```python
dft.transpose()[['AALR3.SA', 'ABCB4.SA']].plot();
```

```python
from sklearn.manifold import TSNE
```

```python
dfc = pd.DataFrame(TSNE(n_components=2).fit_transform(dft), index=dft.index)
dfc.columns = ['x', 'y']
dfc.reset_index(inplace=True)
dfc = dfc.merge(clusters)
```

```python
dfc.plot.scatter(x='x', y='y', c='cluster', colormap='Set1');
```
